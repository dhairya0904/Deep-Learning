# Style Transfer

This notebook deals with style transfer with the help of pretrained vgg16 model. 
Vgg16 is a convolution network for image classification tasks.
Here we will not use the fully connected layers. Instead we will use the intermediate hidden layers to calculate content and style loss.
Content loss is for the original image.
Style loss is for the style image whose style we want to transfer to given image.

Our main task is finding an image that simultaneously matches the content representation of the photograph and the style representation of the respective piece of art.

```python
import time
from PIL import Image
import numpy as np

import keras.backend as K
from keras.applications.vgg16 import VGG16,preprocess_input
from keras.models import Model

from scipy.optimize import fmin_l_bfgs_b
from scipy.misc import imsave
```

```python
height = 512
width = 512

#content image
content_image_path = "winter-wolf.jpg"
content_image = Image.open(content_image_path)
content_image = content_image.resize((width,height))
content_image
```

```python
#style image 

style_image_path = 'blue-moon-lake.jpg'
style_image = Image.open(style_image_path)
style_image = style_image.resize((width, height))
style_image
```

```python
# we will add another dimension to image array so that it can be concatenated 

content_array = np.asarray(content_image,dtype="float32")
content_array = np.expand_dims(content_array,axis=0)
content_array = preprocess_input(content_array)
print(content_array.shape)

style_array = np.asarray(style_image, dtype='float32')
style_array = np.expand_dims(style_array, axis=0)
style_array = preprocess_input(style_array)
print(style_array.shape)
```

# Now we start making our computation graph

```python
content_image = K.variable(content_array)
style_image = K.variable(style_array)
combination_image = K.placeholder((1, height, width, 3))
```

```python
input_tensor = K.concatenate( [content_image, style_image, combination_image], axis = 0 )
```

```python
# we are going to use pre trained vgg16 model for this task

model = VGG16(input_tensor=input_tensor, weights='imagenet',
              include_top=False)
```

```python
#all the layers of vgg16 model
layers = dict((layer.name,layer.output) for layer in model.layers)
layers
```

```python
#this variable can be played with to get different results

content_weight = 0.025
style_weight = 5.0
total_variation_weight = 1.0
```

```python
#initializing loss to 0

loss = K.variable(0.)
```

# Content Loss

For content loss we will use shallow convolution layer because they contain the better stucture features for content image. Shallow layer also guarantees that generated image wil have better structural silimarity with content image.

We will use the hidden layer values of generated combination image and content image to get the loss. Loss wil be equal to euclidean distance between both the images.

```python
#defining the content loss

def content_loss(content,combination):
    return K.sum(K.square(content-combination))

layer_features = layers['block2_conv2']
content_image_features = layer_features[0, :, :, :]
combination_features = layer_features[2, :, :, :]

loss += content_weight * content_loss(content_image_features,
                              combination_features)
```

# Style Loss

Gram matrix is simply the dot product of matrix with its own transpose. It contains the information regaeding style of image. If two images have same gram matrix, it means they have same style but not necessarily content.

Style loss is euclidean distance between the gram matrix of feature map of style image and gram matrix of feature map of generated image.

```python
# for creating the gram matrix

def gram_matrix(x):
    features = K.batch_flatten(K.permute_dimensions(x, (2, 0, 1)))
    gram = K.dot(features, K.transpose(features))
    return gram
```

```python
#defining the style loss

def style_loss(style, combination):
    S = gram_matrix(style)
    C = gram_matrix(combination)
    channels = 3
    size = width * height
    
    return 1.0 * K.sum( K.square(S-C) ) / (4 * (channels ** 2) * (size ** 2))    
```

```python
feature_layers = ['block1_conv2', 'block2_conv2',
                  'block3_conv3', 'block4_conv3',
                  'block5_conv3']

for layer_name in feature_layers:
    layer_features = layers[layer_name]
    style_features = layer_features[1, :, :, :]
    combination_features = layer_features[2, :, :, :]
    sl = style_loss(style_features, combination_features)
    loss += (style_weight / len(feature_layers)) * sl
```

# Total variation loss

It makes the spatial representation of image more smooth. It removes the noise from the image|

```python
def total_variation_loss(x):
    a = K.square(x[:, :height-1, :width-1, :] - x[:, 1:, :width-1, :])
    b = K.square(x[:, :height-1, :width-1, :] - x[:, :height-1, 1:, :])
    return K.sum(K.pow(a + b, 1.25))

loss += total_variation_weight * total_variation_loss(combination_image)
```

```python
#Returns the gradients of combination_image w.r.t. loss.
grads = K.gradients(loss, combination_image)
```

```python


outputs = [loss]
print len(outputs)

#Keras function parameters - input, output
f_outputs = K.function([combination_image], outputs)

#function to get loss and gradient
def eval_loss_and_grads(x):
    x = x.reshape((1, height, width, 3))
    outs = f_outputs([x])
    loss_value = outs[0]
    grad_values = outs[1].flatten().astype('float64')
    return loss_value, grad_values


#evaluator class for optimization
class Evaluator(object):

    def __init__(self):
        self.loss_value = None
        self.grads_values = None

    def loss(self, x):
        assert self.loss_value is None
        loss_value, grad_values = eval_loss_and_grads(x)
        self.loss_value = loss_value
        self.grad_values = grad_values
        return self.loss_value

    def grads(self, x):
        assert self.loss_value is not None
        grad_values = np.copy(self.grad_values)
        self.loss_value = None
        self.grad_values = None
        return grad_values

evaluator = Evaluator()
```

```python
x = np.random.uniform(0, 255, (1, height, width, 3)) - 128.

iterations = 10

for i in range(iterations):
    print('Start of iteration', i)
    start_time = time.time()
    x, min_val, info = fmin_l_bfgs_b(evaluator.loss, x.flatten(),
                                     fprime=evaluator.grads, maxfun=20)
    print('Current loss value:', min_val)
    end_time = time.time()
    print('Iteration %d completed in %ds' % (i, end_time - start_time))
```
